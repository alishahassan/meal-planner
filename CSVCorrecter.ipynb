```python
import csv
import re

# need this for cleaning Allergens and Ingredients fields
def clean_ingredients(text):
    # Replace bullet-like sequences with a semicolon
    text = re.sub(r'(â€¢|•|Â·)\s*', '; ', text)

    # Remove double semicolons or leading semicolon
    text = re.sub(r';\s*;', ';', text).strip()

    # Remove a leading semicolon if it exists
    if text.startswith(';'):
        text = text[1:].strip()

    return text

def remove_bad_encoding(text):
    return text.replace("Â", "")

def second_remove_bad_encoding(text):
    # Fix common fraction encoding issues
    text = text.replace("â…—", "1/8")
    text = text.replace("Â¼", "1/4")
    text = text.replace("Â½", "1/2")
    text = text.replace("Â¾", "3/4")
    text = text.replace("â…›", "1/3")
    text = text.replace("â…œ", "2/3")
    text = text.replace("â…", "1/5")
    text = text.replace("â…–", "2/5")
    text = text.replace("â…•", "3/5")
    text = text.replace("â…", "4/5")
    text = text.replace("â…™", "1/6")
    text = text.replace("â…š", "5/6")
    return text

def remove_contains(text):
    # Remove patterns like '(Contains:  Milk)' or '(Contains: Tree Nuts)'
    text = re.sub(r'\(Contains:\s*[^)]+\)', '', text, flags=re.IGNORECASE).strip()
    return text

everyLine = []

with open('recipes_a.csv', mode='r') as file:
    csvFile = csv.reader(file)
    for line in csvFile:
        templist = []
        for i in range(14):
            if i != 1 and i != 2:
                if i == 3:
                    data = line[i]
                    result = clean_ingredients(data)
                    templist.append(result)
                elif i == 4:
                    data = line[i]
                    result = re.sub(r'([A-Za-z])(\d+)\s*kcal', r'\1 \2kcal', data)
                    templist.append(result)
                elif i == 5:
                    data = line[i]
                    result = re.sub(r'([A-Za-z])(\d+)\s*g', r'\1 \2g', data)
                    templist.append(result)
                elif i == 6:
                    data = line[i]
                    result = re.sub(r'([A-Za-z])(\d+)\s*([a-zA-Z]+)', r'\1 \2\3', data)
                    templist.append(result)
                elif i == 7:
                    data = line[i]
                    result = re.sub(r'([A-Za-z])(\d+)\s*g', r'\1 \2g', data)
                    templist.append(result)
                elif i == 8:
                    data = line[i]
                    result = re.sub(r'([A-Za-z])(\d+)\s*g', r'\1 \2g', data)
                    templist.append(result)
                elif i == 9:
                    data = line[i]
                    result = re.sub(r'([A-Za-z])(\d+)\s*([a-zA-Z]+)', r'\1 \2\3', data)
                    templist.append(result)
                elif i == 10:
                    data = line[i]
                    result = re.sub(r'([A-Za-z])(\d+)\s*g', r'\1 \2g', data)
                    templist.append(result)
                elif i == 11:
                    data = line[i]
                    result = re.sub(r'([A-Za-z])(\d+)\s*mg', r'\1 \2mg', data)
                    templist.append(result)
                elif i == 12:
                    data = line[i]
                    result = re.sub(r'([A-Za-z])(\d+)\s*mg', r'\1 \2mg', data)
                    templist.append(result)
                else:
                    # ingredients list
                    data = line[i]
                    result = clean_ingredients(data)
                    result = remove_bad_encoding(result)
                    result = second_remove_bad_encoding(result)
                    result = remove_contains(result)
                    templist.append(result)
        everyLine.append(templist)
```

```python
# Write the processed data to a new CSV file
with open('new_a.csv', mode='w', newline='', encoding='utf-8') as file:
    csvWriter = csv.writer(file)
    for line in everyLine:
        csvWriter.writerow(line)
```
